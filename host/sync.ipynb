## Imports and constants

[`sync.py`](file_editor_?fp=%2Fapp%2Fdata%2F1iis%2Fsrvbox%2Fhost%2Fsync.py) starts with stdlib-only imports and a repository root constant. The MVP should stay dependency-free so a fresh Ubuntu/Debian host can run it as soon as `python3` exists.

```
#| export
from __future__ import annotations
import argparse
import os
import platform
import shutil
import subprocess
import sys
from dataclasses import dataclass
from pathlib import Path
from typing import Callable
```

```
#| export
ROOT = Path(__file__).resolve().parents[1]
```

## Small logging and command helpers

Keep process helpers tiny and explicit. `log` gives predictable operator output; `run` centralizes subprocess execution so later reconciliation steps can be readable and auditable.

```
#| export
def log(msg: str) -> None:
    print(f"==> {msg}")
```

```
#| export
def run(cmd: list[str], *, check: bool = True) -> subprocess.CompletedProcess[str]:
    log("$ " + " ".join(cmd))
    return subprocess.run(cmd, check=check, text=True)
```

## Guard rails

`apply` is allowed to change the host, so it must run as root and only on the supported OS family. `status` stays cheap and non-root so we can run it locally and in smoke tests.

```
#| export
def require_root() -> None:
    if os.geteuid() != 0:
        raise SystemExit("error: sync.py must run as root")
```

```
#| export
def require_ubuntuish() -> None:
    if not Path("/etc/os-release").exists():
        raise SystemExit("error: /etc/os-release not found")

    data = Path("/etc/os-release").read_text()
    if "Ubuntu" not in data and "Debian" not in data:
        raise SystemExit("error: only Ubuntu/Debian-like hosts are supported for now")
```

## Step model

The MVP uses a tiny `Step` dataclass instead of a framework. Each step has a human-readable name and a nullary function. This keeps the reconciler easy to extend while preserving a clear execution order.

```
#| export
@dataclass(frozen=True)
class Step:
    name: str
    fn: Callable[[], None]
```

## Phase 1 placeholders

These functions define the first visible shape of the reconciler without changing the host yet. Each placeholder names one future idempotent operation, letting `setup.sh` exercise the full end-to-end path immediately.

```
#| export
def ensure_base_dirs() -> None:
    for path in [
        Path("/opt/1iis"),
        Path("/etc/1iis"),
        Path("/var/lib/1iis"),
        Path("/var/log/1iis"),
    ]:
        path.mkdir(parents=True, exist_ok=True)
        shutil.chown(path, user="root", group="root")
        path.chmod(0o755)
        log(f"ensured directory: {path}")
```

## Status command

`status` reports basic facts without mutating the machine. This gives us a safe smoke test for local development, CI-like checks, and quick VM diagnostics before running `apply`.

```
#| export
def check_status() -> None:
    log(f"repo root: {ROOT}")
    log(f"python: {sys.version.split()[0]}")
    log(f"platform: {platform.platform()}")
    log(f"git: {shutil.which('git') or 'missing'}")
    log("status check complete")
```

## Apply command

`apply` is the mutating path. It first checks guard rails, then runs steps in the deliberately boring order: base filesystem shape, identity, access hardening, rate limiting, unattended security updates, and optional ingress.

```
#| export
STEPS = [
    Step("base directories", ensure_base_dirs),
    Step("admin/service users", ensure_admin_user),
    Step("ssh hardening", configure_ssh),
    Step("firewall", configure_firewall),
    Step("fail2ban", configure_fail2ban),
    Step("unattended upgrades", configure_unattended_upgrades),
    Step("caddy", configure_caddy),
]
```

```
#| export
def apply() -> None:
    require_root()
    require_ubuntuish()

    for step in STEPS:
        log(f"Step: {step.name}")
        step.fn()
```

## CLI

The command-line interface is intentionally tiny: `status` for safe observation, `apply` for reconciliation. Keeping the public surface this small helps the shell bootstrap stay stable.

```
#| export
def parse_args() -> argparse.Namespace:
    parser = argparse.ArgumentParser(description="srvbox host reconciler")
    parser.add_argument(
        "command",
        choices=["status", "apply"],
        help="status checks the host; apply reconciles desired state",
    )
    return parser.parse_args()
```

```
#| export
def main() -> None:
    args = parse_args()

    if args.command == "status":
        check_status()
    elif args.command == "apply":
        apply()
    else:
        raise SystemExit(f"unknown command: {args.command}")
```

```
#| export
if __name__ == "__main__":
    main()
```
