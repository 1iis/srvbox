# [1iis/srvbox](https://github.com/1iis/srvbox)<br>Server Baseline CRAFT

> A small production server setup, in Python.

`srvbox` is a minimal, hardened, boring production server baseline.  
If [`devbox`](https://github.com/1iis/devbox) is the comfortable SWE workstation, `srvbox` is the lean production substrate.

It provisions the host into a known baseline, then gets out of the way.

## Current status

`srvbox` has reached its first MVP: a small, artifact-deployed, hardened production server baseline for Ubuntu/Debian-like hosts.

The deployment spine is:

```text
local repo
→ scripts/deploy.sh
→ tar artifact + generated runner
→ scp to /tmp/deploy/1iis/
→ ssh -t user@host "sudo sh runner"
→ scripts/setup.sh
→ /opt/1iis/srvbox/releases/<timestamp>/
→ /opt/1iis/srvbox/current
→ host/sync.py apply
```

The baseline now includes:

- root-owned timestamped releases under `/opt/1iis/srvbox/releases/`
- read-only `status` and root-only `apply` reconciler commands
- base `/opt/1iis`, `/etc/1iis`, `/var/lib/1iis`, `/var/log/1iis` directories
- explicit access policy: human users unmanaged; deploy user pre-exists with SSH + sudo
- SSH hardening via validated `/etc/ssh/sshd_config.d/` drop-in
- `ufw` firewall baseline with SSH preserved
- fail2ban installed/configured, intentionally inactive during ingress/TLS bootstrap
- unattended security upgrades ensured
- optional Caddy ingress using the official Caddy apt repository
- global Caddyfile importing app fragments from `/etc/caddy/sites.d/*.caddy`

Validated on Ubuntu Server through deploy, harden, reboot, redeploy, and smoke checks.

Installed `srvbox` release trees are owned by `root:root`; `srvbox` is infrastructure code executed as root.

### `scripts/deploy.sh`

Workstation-side deployment command.

Usage:

```sh
./scripts/deploy.sh -s /path/to/repo -d user@host
```

Defaults:

```text
-s / --src = .
-d / --dst = required SSH destination
DEPLOY_ROOT = /tmp/deploy/1iis
```

It:

1. creates a tarball artifact of the repo
2. excludes dev/cache files
3. normalizes tar ownership with `--owner=0 --group=0 --numeric-owner`
4. writes a temporary remote runner script
5. uploads artifact + runner to the target server
6. runs the runner with `ssh -t ... sudo ...`

### `host/sync.py`

Stdlib-only Python reconciler exported from the notebook [`host/sync`](dialog_?name=1iis%2Fsrvbox%2Fhost%2Fsync).

CLI:

```text
status

```

`status` is read-only and works without root. It reports host facts for the baseline: OS, root state, SSH, firewall, fail2ban, unattended upgrades, and Caddy.

`apply` requires root and converges the host through explicit idempotent steps:

```text
base directories
access policy
ssh hardening
firewall

```

Key behavior:

- base `/opt/1iis`, `/etc/1iis`, `/var/lib/1iis`, `/var/log/1iis` dirs are ensured as `root:root 0755`
- human users are unmanaged; deploy access must already exist via SSH + sudo
- no `srvbox` daemon/service user exists
- SSH hardening is written as `/etc/ssh/sshd_config.d/90-1iis-srvbox.conf` and validated with `sshd -t`
- `ufw` denies incoming, allows outgoing, and allows SSH before enabling
- fail2ban is installed/configured but disabled by policy during ingress/TLS bootstrap
- unattended upgrades are installed/configured with apt timers enabled
- Caddy ingress is policy-controlled by `ENABLE_CADDY`
- Caddy uses the official Caddy apt repository
- app-owned Caddy fragments live in `/etc/caddy/sites.d/*.caddy`
- apt operations retry to tolerate package-manager lock races

## Key design decisions

### No `git` on production servers

Production servers do not need repository access, GitHub credentials, deploy keys, or even `git`.

Deployments are artifact-based:

```text
tar + scp + ssh + sudo + sh
```

This keeps the production surface small.

### Use uploaded runner file, not SSH here-doc into sudo

Do **not** do this:

```sh
ssh -t user@host "sudo sh -s" <<'EOF'
...
EOF
```

If `sudo` prompts for a password, it may consume the script body as password input.

Correct pattern:

```text
1. upload artifact
2. upload generated runner script
3. ssh -t user@host "sudo sh /tmp/deploy/1iis/repo.run.sh"
```

This separates:

```text
TTY/password input
```

from:

```text
script payload
```

### Root owns infrastructure code

`srvbox` release code under `/opt` is root-owned because `srvbox` is run as root.

Human deploy users may trigger deployment via `sudo`, but they should not own installed production code.

### No `srvbox` service user yet

`srvbox` is currently a reconciler, not a daemon.

If we later add a long-running `srvbox-agent`, create a system user then.

### Future app identity model

Future apps should use per-app system users:

```text
/opt/1iis/<app>/releases/...   root:root
/opt/1iis/<app>/current        root-managed symlink
/etc/1iis/<app>/               root-owned config/secrets
/var/lib/1iis/<app>/           app-user:app-user
/var/log/1iis/<app>/           app-user:app-user
```

systemd services should run as:

```ini
User=iis-<app>
Group=iis-<app>
WorkingDirectory=/opt/1iis/<app>/current
EnvironmentFile=/etc/1iis/<app>/env
```

The app should be able to read its deployed code, but not modify it.

### Caddy as optional host ingress

Caddy belongs to `srvbox` as an optional-but-first-class host capability, not as app/domain logic.

`ENABLE_CADDY = True` means `srvbox` owns the baseline HTTP ingress substrate. `ENABLE_CADDY = False` means public HTTP ingress is closed by policy.

Apps own routing fragments under `/etc/caddy/sites.d/*.caddy`; higher infra owns DNS, VPS provisioning, and secret materialization.

### Phase 3 — Caddy ingress ✅

Caddy is the optional-but-first-class HTTP ingress layer.

When `ENABLE_CADDY = True`, `srvbox`:

- installs Caddy from the official Caddy apt repository
- owns `/etc/caddy/Caddyfile`
- creates `/etc/caddy/sites.d/`
- imports app-owned fragments from `/etc/caddy/sites.d/*.caddy`
- validates Caddy config before service changes
- opens `80/tcp` and `443/tcp` in `ufw`
- enables/starts or reloads `caddy.service`

When `ENABLE_CADDY = False`, public HTTP ingress is closed by policy.

Caddy deliberately contains no app/domain intent. Apps deploy their own fragments; higher infra may provision DNS/VPS/secrets and call the same deploy primitive.

Contract:

```text
srvbox owns host ingress substrate
apps own routing intent
infra owns orchestration
```

This keeps standalone deploys and future infra automation on the same path.

### Next chapter — App deployment and infra integration

`srvbox` should support three operating modes with the same underlying contracts:

1. **Standalone host** — a repo deploy can harden a VPS and run an app with only SSH, sudo, DNS pointed at the host, and app-provided config.
2. **Infra-orchestrated host** — higher infra provisions VPS/DNS/secrets/storage, then calls the same `srvbox` deploy primitive.
3. **Private/internal host** — services may run behind WireGuard/Tailnet/private networking with no public HTTP ingress.

The important invariant: manual and automated paths should be the same path. Infra adds orchestration around `srvbox`; it should not replace the local deploy/reconcile primitive.

Expected app-layer shape:

- first app deploys copy the `srvbox` artifact/release pattern manually
- abstraction is extracted only after repetition hurts
- app code releases remain root-owned under `/opt/1iis/<app>/releases/`
- `/opt/1iis/<app>/current` is a root-managed symlink
- per-app services run as system users named `iis-<app>`
- app config/secrets live under `/etc/1iis/<app>/`
- app state lives under `/var/lib/1iis/<app>/`
- app logs live under `/var/log/1iis/<app>/`
- systemd units point at the app's current release and use an environment file
- public HTTP apps write Caddy fragments under `/etc/caddy/sites.d/*.caddy`
- private/internal services may skip Caddy entirely

The first concrete app target is the standalone `1i.is` website deployment.
